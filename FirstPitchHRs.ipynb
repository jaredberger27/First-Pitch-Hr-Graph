```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('firstpitch.csv')
```

```python
df.head(10)
```

```python
df.columns
```

```python
df.describe
```

```python
df.loc['mean_fruit'] = df.mean()
```

```python
df.head(25)
```

```python
palette = ['red', 'blue', 'yellow']
fig = plt.figure(figsize=(10,10))
fig = sns.scatterplot(data=df, x= 'plate_x', y = 'plate_z', hue = 'pitch_type', s=80, 
                      edgecolor ='black', palette = palette)
fig.set_aspect('equal')
strikezone = plt.Rectangle(((-(17/2)/12),1.5), 17/12,2,
                         fill=None, ec="black", zorder=3, alpha=.7, lw=5)
plt.gca().add_patch(strikezone)
sns.despine(left= True, bottom=True)
plt.xlabel('')
plt.ylabel('')

```
